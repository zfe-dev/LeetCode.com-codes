Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.



```python
from typing import Optional

class TreeNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    ######### Wrong exercice, returns max of tree, not depth to go to maximum value
    """def maxDepth(self, root: Optional[TreeNode]) -> int:

        if type(root.left) is TreeNode and type(root.right) is TreeNode:
            return max(self.maxDepth(root.left), self.maxDepth(root.right))
        elif type(root.left) is TreeNode:
            return max(self.maxDepth(root.left), root.val if root.right == None else max(root.val, root.right))
        elif type(root.right) is TreeNode:
            return max(self.maxDepth(root.right), root.val if root.left == None else max(root.val, root.left))
        elif type(root.left) is int and root.right == None:
            return max(root.val, root.left)
        elif type(root.right) is int and root.left == None:
            return max(root.val, root.right)
        else:
            return root.val"""
    
    ######### Good solution, returns depth to go to the maximum value
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if root == None:
            return 0
        
        left_node_depth = self.maxDepth(root.left)
        right_node_depth = self.maxDepth(root.right)
        
        return 1 + max(left_node_depth, right_node_depth)
            
            

print("depth: ", Solution().maxDepth(TreeNode(3, TreeNode(9), TreeNode(20, TreeNode(15), TreeNode(7)))))
#print("depth: ", Solution().maxDepth(TreeNode(1, None, TreeNode(2))))
```
